```python
import pdfplumber
import math

def open(file):
    def getVndData(**pos_vnd):
        finalPrice = get_text(p0,pos_vnd['finalPrice'])
        if finalPrice != None:
            finalPrice= finalPrice.replace('$', '')
        
        d = {
            'sn':get_text(p0,pos_vnd['sn']),
            'id':get_text(p0,pos_vnd['id']),
            'name':get_text(p0,pos_vnd['name']).replace(' 新廠商', '').replace('停詢', ''),
            'type':get_text(p0,pos_vnd['type']),
            'quo':get_text(p0,pos_vnd['quo']).replace('$', ''),
            'quoRate':get_text(p0,pos_vnd['quoRate']),
            'finalPrice':finalPrice,
            'finalRate':get_text(p0,pos_vnd['finalRate']),
            'bargain_times':get_text(p0, pos_vnd['bargain_times']),
            'level':get_text(p0, pos_vnd['level']),
        }
        
      
        return d
    
    pdf_1 = pdfplumber.open(file)
    p0 = pdf_1.pages[0]

    #--------------------個欄位資料位置-------------------------------------------------------------
    pos_caseID = {'x0':22.320, 'top':91.744, 'x1':70.320, 'bottom':109.600}
    pos_caseName = {'x0':28.280, 'top':135.104, 'x1':393.600, 'bottom':152.960}
    pos_sn = {'x0':81.320, 'top':91.744, 'x1':93.320, 'bottom':109.600}
    pos_comp = {'x0':100.360, 'top':91.744, 'x1': 118.360, 'bottom':109.600}
    pos_budget = {'x0':700.750, 'top':91.304, 'x1': 815.87, 'bottom':109.16}
    pos_type = {'x0':148.960, 'top':88.744, 'x1': 226.960, 'bottom':115.600}
    pos_currency ={'x0': 715.750, 'top': 49.304, 'x1':751.750, 'bottom':67.160}
    pos_category = {'x0':252.430, 'top':91.904, 'x1': 350.160, 'bottom':109.760}
    pos_level = {'x0':332.230, 'top':134.504, 'x1': 338.230, 'bottom':152.360}
    pos_require_rate = {'x0':682.840, 'top':376.984, 'x1':688.840, 'bottom':394.840}
    
    pos_vendorID= {'x0':46.790, 'top':193.628, 'x1':127.790, 'bottom':207.020}
    pos_vendorName= {'x0':46.720, 'top':210.052, 'x1':185.400, 'bottom':226.420}
    pos_vendorType={'x0':165.400, 'top':193.080, 'x1':185.400, 'bottom':226.420}
    pos_quo = {'x0':361.800, 'top':190.680, 'x1': 439.800, 'bottom':207.960}
    pos_quoRate = {'x0':420.360, 'top':216.440, 'x1': 439.800, 'bottom':231.320}
    pos_finalPrice = {'x0':635, 'top':190.680, 'x1':725.680 , 'bottom':205.560}
    pos_finalRate = {'x0':708.140, 'top':215.840, 'x1':725.140 , 'bottom':230.720}
    pos_vendorSn = {'x0':20.400, 'top':193.628, 'x1':44.400, 'bottom':215}
    pos_bargain_times = {'x0':448.640, 'top':190.344, 'x1': 454.640, 'bottom':208.200}
    pos_vnd_level = {'x0':142.200, 'top':193.080, 'x1': 152.200, 'bottom':207.960}
    #----------------------------------------------------------------------------------------------
    
    #-------擷取pdf內案件資料
    caseID = get_text(p0,pos_caseID)
    caseName = get_text(p0,pos_caseName)
    sn = get_text(p0,pos_sn)
    comp = get_text(p0,pos_comp)
    budget = get_text(p0,pos_budget).replace('$', '')
    caseType = get_text(p0,pos_type)
    currency = get_text(p0,pos_currency)
    category = get_text(p0, pos_category)
    level = get_text(p0, pos_level)
    require_rate = get_text(p0, pos_require_rate)
    
    #-------擷取pdf內報價廠商資料,gap是每個廠商的間距
    vnd = []
    gap_id = [0, 36.48, 36.48, 36.44, 36.48]  #適用vndId,finalrate
    gap_name = [0, 36.44, 36.48, 36.48, 36.44] #適用vndname, quo , vendortype, finalprice, bargain_times, vnd_level
    gap_rate = [0, 36.48, 36.44, 36.48, 36.48] #適用quorate, vndsn, 
    
    i = 0   #報價廠商次序
    j = 1   #呈核表頁次
        
    while True:
        if i == 5 and j == len(pdf_1.pages):#
            break
        elif i == 5 and j != len(pdf_1.pages):
            i = 0
            j += 1
            p0 = pdf_1.pages[j-1]
            pos_vendorID= {'x0':46.790, 'top':193.628, 'x1':127.790, 'bottom':207.020}
            pos_vendorName= {'x0':46.720, 'top':210.052, 'x1':185.400, 'bottom':226.420}
            pos_vendorType={'x0':165.400, 'top':193.080, 'x1':185.400, 'bottom':226.420}
            pos_quo = {'x0':361.800, 'top':190.680, 'x1': 439.800, 'bottom':207.960}
            pos_quoRate = {'x0':420.360, 'top':216.440, 'x1': 439.800, 'bottom':231.320}
            pos_finalPrice = {'x0':635, 'top':190.680, 'x1':725.680 , 'bottom':205.560}
            pos_finalRate = {'x0':710.140, 'top':215.840, 'x1':724.140 , 'bottom':230.720}
            pos_vendorSn = {'x0':20.400, 'top':193.628, 'x1':44.400, 'bottom':215}
            pos_bargain_times = {'x0':448.640, 'top':190.344, 'x1': 454.640, 'bottom':208.200}
            pos_vnd_level = {'x0':142.200, 'top':193.080, 'x1': 152.200, 'bottom':207.960}
        
        pos_vendorID['top'] += gap_id[i]
        pos_vendorID['bottom'] += +gap_id[i]    
        pos_vendorName['top']+=gap_name[i]
        pos_vendorName['bottom']+=gap_name[i]
        pos_vendorType['top']+=gap_name[i]
        pos_vendorType['bottom']+=gap_name[i]
        pos_quo['top']+=gap_name[i]
        pos_quo['bottom']+=gap_name[i]
        pos_finalPrice['top']+=gap_name[i]
        pos_finalPrice['bottom']+=gap_name[i]
        pos_vendorSn['top']+=gap_rate[i]
        pos_vendorSn['bottom']+=gap_rate[i]
        pos_quoRate['top']+=gap_rate[i]
        pos_quoRate['bottom']+=gap_rate[i]
        pos_finalRate['top']+=gap_id[i]
        pos_finalRate['bottom']+=gap_id[i]
        pos_bargain_times['top']+=gap_name[i]
        pos_bargain_times['bottom']+=gap_name[i]
        pos_vnd_level['top']+=gap_name[i]
        pos_vnd_level['bottom']+=gap_name[i]

        vndsn = get_text(p0,pos_vendorSn)
#         print('i', i, 'sn', sn, 'j', j)
        if vndsn == None:
            break
        i += 1

        vnd.append(getVndData(
            sn = pos_vendorSn,
            id = pos_vendorID,
            name = pos_vendorName,
            type = pos_vendorType,
            quo = pos_quo,
            quoRate = pos_quoRate,
            finalPrice = pos_finalPrice,
            finalRate = pos_finalRate,
            bargain_times = pos_bargain_times,
            level = pos_vnd_level
        ))
    
    #將資料變成一個Bidcase物件

    case= Bidcase(caseID, sn, comp, caseName, budget, caseType, currency, category, level, require_rate,\
              vnd)
    del pdf_1
    return case
    
def get_text(pdf, dict_pos):
    crop = pdf.within_bbox((dict_pos['x0'],dict_pos['top'],dict_pos['x1'],dict_pos['bottom']))
    del pdf
    return crop.extract_text()

def round_down(n, decimals=0):#無條件捨去至小數decimals位  
    multiplier = 10 ** decimals
    return int(n * multiplier) / multiplier


class Bidcase():
    def __init__(self, ID, sn, comp, name, budget, caseType, currency, category, level, require_rate, \
                 vnd):
        self.id = ID
        self.sn = sn
        self.comp = str(comp)
        self.name = name
        self.budget = budget
        self.currency = currency
        self.category = category
        self.level = level
        self.vnd = vnd
        self.require_rate = require_rate
        
        
        #vndSn用來計算vnd[]要用第幾筆資料
        self.vndSn = 0
        self.awd_sn = 0 #預設得標廠商序號0
        self.cheapest_sn = self.cheapest()
        
        caseType = caseType.split(' ')[0]
        if self.sn != '00':
            self.type = '變更案'
            self.award_vnd = self.vnd[self.awd_sn]['name']
            n = int(self.budget.replace(',', ''))
            if n >= 0:
                self.change = '追加'
            elif n == 0:
                self.change = '變更'
            else:
                self.change = '追減'
        
        elif caseType =='一般' and self.sn == '00' and vnd[self.awd_sn]['type'] == '預約':
            self.type = '預約子案'
            self.award_vnd = self.vnd[self.awd_sn]['name']
        
        elif caseType =='補辦':
            self.type = '補辦案'
            self.award_vnd = self.vnd[self.awd_sn]['name']
        else:
            self.type = '一般案'
        
        #----------計算議價總次數確定整案是否有議價-------------------------
        sum_bargain_count = 0
        for item in self.vnd:
            sum_bargain_count += int(item['bargain_times'])
        if sum_bargain_count > 0:
            self.is_bargain = True
        else:
            self.is_bargain = False
        
        
    
    def cheapest(self):
        price =[]
        rate = lambda currency, r: 1 if currency == '新台幣' else 1 + int(r)/100
        for item in self.vnd:
            if item['finalPrice'] == None:
                #以稅率判斷稅前價, 新台幣一律為1 稅後金額/(1 + finalRate or quoRate)
                price.append(int(item['quo'].replace(',', ''))/rate(self.currency, item['quoRate']))
            else:
                price.append(int(item['finalPrice'].replace(',', ''))/rate(self.currency, item['finalRate']))
        return price.index(min(price))
    
    #針對沒有議價時較預算差異的計算方式
    def dif_rate(self, vndSn):
        if self.is_bargain:
            n1 = int(self.vnd[vndSn]['finalPrice'].replace(',', ''))
        else:
            n1 = int(self.vnd[vndSn]['quo'].replace(',', ''))

        n2 = int(self.budget.replace(',',''))
        dif = round_down((n1-n2)/n2*100, 2)
        
        if n1 < 0 and n2 < 0:
            dif = '-' + str(dif) + '%'
        elif dif > 0:
            dif = '+' + str(dif) + '%'
        else:
            dif = str(dif) + '%'
        
        return dif
        
    def is_equaltoBudget(self, vndSn):
        if self.is_bargain:
            return self.vnd[vndSn]['finalPrice'].replace(',', '') == self.budget.replace(',', '')
        else:
            return self.vnd[vndSn]['quo'].replace(',', '') == self.budget.replace(',', '')
    
    def is_overBudget(self, vndSn):
        if self.vnd[vndSn]['finalPrice'] == None:
            return int(self.vnd[vndSn]['quo'].replace(',', '')) > int(self.budget.replace(',', ''))
        else:
            return int(self.vnd[vndSn]['finalPrice'].replace(',', ''))  > int(self.budget.replace(',', ''))
        
    def dif_text(self, vndSn):
        if self.is_equaltoBudget(vndSn):
            return '同預算價格'
        else:
            return '較預算' + self.dif_rate(vndSn)
    
        
    def submission_text(self, vndSn, easymode = None):
        #==================initial========================
        prefixCur =''
        suffixCur = '元'
        prefixVnd = ''
        convertCur = ''
        
        if self.currency == '人民幣':
            prefixCur = 'RMB'
            suffixCur = ''
        
        def name_format(var):
            return '「' + str(var) +'」'
        
        def price_format(var):
            return prefixCur + str(var) + suffixCur
        
        if self.vnd[vndSn]['finalPrice'] == None:#沒議價時
            award_price = price_format(self.vnd[vndSn]['quo'])
        else:
            award_price = price_format(self.vnd[vndSn]['finalPrice'])
        
        if easymode == None:
            easymode = False
            
        
        if easymode: #=================簡易版=========================
            if self.vnd[vndSn]['finalPrice'] != None:
                b_prefix = '議後'
            else:
                b_prefix = ''
            
            
            if self.type == '變更案':
                a = '1.本案為' + self.sn + '次變更案。\n'
                b = '2.'+ b_prefix + '擬以' + award_price + '向' + name_format(self.vnd[vndSn]['name'])+ '辦理變更，'\
                + self.dif_text(vndSn) + '。\n'
            
            elif self.type == '預約子案':
                a = '1.本案為預約子案，遴選合約商' + name_format(self.vnd[vndSn]['name']) + '施作。\n'
                b = '2.'+ b_prefix + '擬以' + award_price + '交' + name_format(self.vnd[vndSn]['name'])+ '承攬，'\
                + self.dif_text(vndSn) + '。\n'
            
            elif self.type == '補辦案':
                a = '1.本案為補辦案，已由' + name_format(self.vnd[vndSn]['name']) + '先行施作。\n'
                b = '2.'+ b_prefix + '擬以' + award_price + '交' + name_format(self.vnd[vndSn]['name'])+ '承攬，'\
                + self.dif_text(vndSn) + '。\n'
            else:
                a = '1.本案為' + self.comp + name_format(self.name) + '，預算編列' + price_format(self.budget) + '。\n'
                b = '2.'+ b_prefix + '擬以' + award_price + '交' + name_format(self.vnd[vndSn]['name'])+ '承攬，'\
                + self.dif_text(vndSn) + '。\n'
                
            return a + b
        
        else:#==================三段式=========================
            if self.type == '變更案':
                prefixVnd = '原承商'
                n = int(self.budget.replace(',', ''))
                if n >= 0:
                    goesWay = '追加'
                elif n == 0:
                    goesWay = '變更'
                else:
                    goesWay = '追減'

                part1 = '1.本案為' + self.sn + '次變更案，預算編列'+ price_format(self.budget) + '。\n'
                part3 = '3.擬以' + award_price + '向' + name_format(self.vnd[vndSn]['name']) + '辦理' + goesWay + '。\n'
            elif self.type == '預約子案':
                part1 = '1.本案為預約子案，遴選合約商' + name_format(self.vnd[vndSn]['name']) + '施作。\n'
                part3 = '3.擬以' + award_price +'交由' + name_format(self.vnd[vndSn]['name']) + '承攬。\n'
            elif self.type == '補辦案':
                part1 = '1.本案為補辦案，已由' + name_format(self.vnd[vndSn]['name']) + '先行施作。\n'
                part3 = '3.擬以' + award_price +'交由' + name_format(self.vnd[vndSn]['name']) + '承攬。\n'
            else:
                part1 = '1.本案為' + self.comp + name_format(self.name) + '，預算編列' + price_format(self.budget) + '。\n'
                part3 = ''

            #part2
            if self.vnd[vndSn]['finalPrice'] == None:#沒議價時
                part2 = '2.經洽' + prefixVnd + name_format(self.vnd[vndSn]['name']) + '詢價後為'\
                +price_format(self.vnd[vndSn]['quo']) + '，' + self.dif_text(vndSn) + '。\n'
            else:#有議價時
                part2 = '2.經洽' + prefixVnd + name_format(self.vnd[vndSn]['name']) + '詢議價後由'\
                +price_format(self.vnd[vndSn]['quo']) + '降為' +price_format(self.vnd[vndSn]['finalPrice'])\
                + '，' + self.dif_text(vndSn) + '。\n'


            return part1+ part2 + part3


if __name__ == '__main__':
    file='d:/new.pdf'
    myCase = open(file)
    sn = 0
    
    print('檔案', file)
    print('工號:', myCase.id)
    print('公司:', myCase.comp)
    print('次別:', myCase.sn)
    print('工程名稱:', myCase.name)
    print('幣別:', myCase.currency)
    print('性質:', myCase.type)
    print('詢價專類:', myCase.category)
    print('需求等級:', myCase.level)
    print('需求稅率:', myCase.require_rate)
    print('預算:', myCase.budget)
    print('報價總家數:', len(myCase.vnd))
    print('有無議價:', myCase.is_bargain)
    print('最便宜廠商次序:', myCase.cheapest_sn +1)
    for sn in range(len(myCase.vnd)):
        print('----------------第{x}家報價廠商資料-------------'.format(x = sn+1))
        print('報價次序:', myCase.vnd[sn]['sn'])
        print('廠商:', myCase.vnd[sn]['name'])
        print('廠商編號:', myCase.vnd[sn]['id'])
        print('廠商報價方式:', myCase.vnd[sn]['type'])
        print('廠商等級:', myCase.vnd[sn]['level'])
        print('報價:', myCase.vnd[sn]['quo'])
        print('報價稅率:', myCase.vnd[sn]['quoRate'])
        print('議價次數:', str(myCase.vnd[sn]['bargain_times']))
        print('議後:', str(myCase.vnd[sn]['finalPrice']))
        print('議後稅率:', str(myCase.vnd[sn]['finalRate']))
        print('第{x}家廠商是否超預算:'.format(x = sn+1), myCase.is_overBudget(sn))
        sn += 1

#     print('\n議價說明:')
#     print(myCase.submission_text(sn, easymode = False))
    

        
```
